```python
import pandas as pd

# Load the initial dataset
file_path = 'preprocessed_dataset.csv'
df = pd.read_csv(file_path)
```

```python
# Dropping 'Notebook' and 'Hardware' columns
df_dropped_columns = df.drop(['Notebook', 'Hardware', 'Country', 'ML_Framework','Visualization_tool','Media_sources','Gender'], axis=1)

# Dropping rows where 'Yearly_compensation' is missing
#df_dropped_rows = df_dropped_columns.dropna(subset=['Yearly_compensation'])

# Dropping rows where 'Years_of_machine_learning' is missing
#df_dropped_ml_years = df_dropped_rows.dropna(subset=['Years_of_machine_learning'])
```

```python
dataset=df_dropped_columns.dropna()
dataset
```

```python
dataset['Years_of_machine_learning'] = dataset['Years_of_machine_learning'].replace('MachineLearningEngineer', '0')
dataset = dataset[dataset['Education'] != 'I prefer not to answer']
```

```python
dataset.Job.value_counts()
```

```python
import re

# Function to convert compensation ranges to average values
def convert_range_to_average(compensation_range):
    if "-" in compensation_range:
        start, end = map(int, compensation_range.replace(",", "").replace("$", "").replace("<", "").replace(">", "").split("-"))
        return round((start + end + 1) / 2)
    else:
        return int(compensation_range.replace(",", "").replace("$", "").replace("<", "").replace(">", "").replace(" ", ""))

# Function to extract numbers from strings and calculate the middle value
def extract_and_calculate_middle(s):
    numbers = re.findall(r'\d+', s.replace(",", "").replace("$", "").replace("<", "").replace(">", "").replace(" ", ""))
    if len(numbers) == 2:
        start = int(numbers[0])
        end = int(numbers[1])
        middle = (start + end) // 2
        return middle
    elif len(numbers) == 1:
        return int(numbers[0])
    else:
        return None
# Function to remove text within parentheses
def remove_parentheses(text):
    return re.sub(r'\([^)]*\)', '', text)
```

```python
dataset['ML_algorithm'] = dataset['ML_algorithm'].apply(remove_parentheses)
dataset['Course_platform']=dataset['Course_platform'].apply(remove_parentheses)
dataset['IDE']=dataset['IDE'].apply(remove_parentheses)
dataset['Yearly_compensation'] = dataset['Yearly_compensation'].apply(convert_range_to_average)
dataset['Years_of_programming'] = dataset['Years_of_programming'].apply(extract_and_calculate_middle)
dataset['Years_of_machine_learning'] = dataset['Years_of_machine_learning'].apply(extract_and_calculate_middle)
dataset['Company_size'] = dataset['Company_size'].apply(extract_and_calculate_middle)
dataset['Age'] = dataset['Age'].apply(extract_and_calculate_middle)
dataset
```

```python
df=dataset.copy()
df = df[(df['Job'] != 'DBA/Database Engineer')&(df['Job'] != 'Data Architect')&(df['Job'] != 'Developer Relations/Advocacy')&(df['Job'] != 'Developer Advocate')&(df['Job'] != 'Data Administrator')&(df['Job'] != 'Engineer (non-software)')&(df['Job'] != 'Teacher / professor')&(df['Job'] != 'Business Analyst')&(df['Job'] != 'Data Engineer')&(df['Job'] != 'Product/ProjectManager')&(df['Job'] != 'Statistician')]
df.Job.value_counts()
```

```python
# Exporting the cleaned dataset
output_file_path = 'final_dataset.csv'
df.to_csv(output_file_path, index=False)
```
